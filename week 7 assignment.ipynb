```python
## importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
## Loading the dataset 

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
## group by species and calculate mean 

iris_df.groupby('species').mean()
```

### Observations:
- Setosa has the smallest petals but wider sepals.
- Virginica has the largest petals on average.

## Line chart

```python
plt.figure(figsize=(8, 4))
for feature in iris.feature_names[:3]:  # First 3 features
    plt.plot(iris_df[feature][:50], label=feature)  # First 50 samples
plt.title('Feature Trends (First 50 Samples)')
plt.xlabel('Sample Index')
plt.ylabel('Measurement (cm)')
plt.legend()
plt.show()
```

## Bar chart

```python
iris_df.groupby('species')['sepal length (cm)'].mean().plot(kind='bar')
plt.title('Average Sepal Length by Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

## Histogram

```python
iris_df['petal length (cm)'].hist(bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

## Scatter plot

```python
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df)
plt.title('Sepal vs Petal Length by Species')
plt.show()
```
